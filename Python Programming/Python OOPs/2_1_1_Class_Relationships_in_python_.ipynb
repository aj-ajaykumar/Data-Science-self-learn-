```python
# Parent class
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def display(self):
        print(f"Name: {self.name}, Age: {self.age}")

# Child class
class Student(Person):
    def __init__(self, name, age, student_id):
        # Call the constructor of the parent class
        super().__init__(name, age)
        self.student_id = student_id

    def display(self):
        # Call the display method of the parent class
        super().display()
        print(f"Student ID: {self.student_id}")

# Create an object of the Student class
student = Student("Alice", 20, "S12345")
student.display()
# we will learn more about inherutance later
```

**Aggrigation with getters ans setters**



**Problem Statement**

Once we have appropriate accessor and mutator methods we can start accessing the private variables of the aggregated class using those methods.

```python
class customer:
  def __init__(self,name,age,phone_number,address):
    self.name = name
    self.age = age
    self.phone_number = phone_number
    self.address = address

  #methods
  def view_details(self):
    print("Name : ",self.name, ",age : ",self.age," ,phone Number : ",self.phone_number)
    Address.show_address(self.address) # use method from Address class

  def update_details(self,name,age,phone_number,address):
    self.name = name
    self.age = age
    self.phone_number = phone_number
    self.address = address




class Address:
  def __init__(self,door_no,street,pincode):
    # Private Attributes
    self.__door_no = door_no
    self.__street = street
    self.__pincode = pincode

  # getters
  @property
  def door_no(self):
    return self.__door_no
  @property
  def street(self):
    return self.__street
  @property
  def pincode(self):
    return self.__pincode

  #setters
  @door_no.setter
  def door_no(self,door_no):
    self.__door_no = door_no
  @street.setter
  def street(self,street):
    self.__street = street
  @pincode.setter
  def pincode(self,pincode):
    self.__pincode = pincode



  def show_address(self):
    print("door_no : ",self.door_no, ",street : ",self.street," ,pincode : ",self.pincode)

  def update_address(self,door_no,street,pincode):
    self.door_no = door_no
    self.street = street
    self.pincode = pincode


add1 = Address("3-832","mail road",53002)
cust1 = customer("ajay",35,"16626374",add1)

cust1.view_details()
print("\n Updating..... tha Address")
add1.update_address("45-33","mallareddy streert",538646)
cust1.view_details()
print("\n calling.... Show Address Method")
add1.show_address()

```

**Dependency via Formal Parameter**



**Problem Statement**

Sometimes a class may depend on another class for some of its use. This is not a strict relationship and hence won’t appear in the class diagram. For example, in the below code, the Customer class depends on a payment object for purchasing. Here payment is a local variable and not an attribute.

we can also use the one calss variable in an another calss methods directly using calss name



```python
class Customer:
    def __init__(self, name, age, phone_no):
        self.name = name
        self.age = age
        self.phone_no = phone_no

    def purchase(self, payment):
        if payment.type == "card":
            print ("Paying by card")
        elif payment.type == "e-wallet":
            print ("Paying by wallet")
        else:
            print ("Paying by cash")

class Payment:
    def __init__(self, type):
        self.type = type

payment1=Payment("card")
cust=Customer("Ajay",25,1234)

cust.purchase(payment1)
```

>**Note:**

>**composition with encapsulation**

```python
# encapsulation with getters and setters & encapsulation
class salary:
  def __init__(self,pay,bonus):
    self.__pay = pay
    self.__bonus = bonus

  # getters
  @property
  def pay(self):
    return self.__pay
  @property
  def bonus(self):
    return self.__bonus

  # setters
  @pay.setter
  def pay(self,pay):
    self.__pay = pay
  @bonus.setter
  def bonus(self,bonus):
    self.__bonus = bonus

  #method
  def calculate_salary(self):
    return self.pay*12 + self.bonus


class employee:
  def __init__(self,name,age,pay,bonus):
    self.name =  name
    self.age = age
    self.salary_obj = salary(pay,bonus) # here we creating object directly inside the class.

  def anuual_Salary(self):
    print(self.name,"annual salary is : ",self.salary_obj.calculate_salary())


emp1 = employee("Ajay",25,40000,30000)
emp1.anuual_Salary()
# here we access the salary method from employee class
print(emp1.salary_obj.calculate_salary())
# here we can't access the salary onject without employee class
```
