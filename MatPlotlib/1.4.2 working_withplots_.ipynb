# Plotting Histogram

```python
data = np.random.randn(1000)
plt.hist(data,bins=30,color='cyan',edgecolor='red')


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

```python
# with kernal density estimater
import seaborn as sns
data = np.random.randn(1000)
sns.histplot(data,bins=30,kde=True,color='cyan',edgecolor='red')


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')
plt.show()
```

> **Multiple Histograms with Subplots**

```python
data1 = np.random.randn(1000)
data2 = np.random.normal(loc=3, scale=1, size=1000)

# subplots
fig,ax = plt.subplots(nrows=1,ncols=2,figsize=(12,4))

ax[0].hist(data1,bins=20,color='cyan',edgecolor='r')
ax[0].set_title("Histogram 1")

ax[1].hist(data2,bins=20,color='yellow',edgecolor='black')
ax[1].set_title("Histogram 2")

plt.grid(linestyle=':',color='gray')
for a in ax:
  a.set_xlabel("Value")
  a.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

> **Stacked Histogram using Matplotlib**

```python
sns.histplot([data1,data2],bins=30,color=['cyan','hotpink'])


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')

plt.legend(['dataset1','dataset2'])
plt.show()
```

```python
# plt not contain kde and sns not contain stacked

plt.hist([data1,data2],bins=30,stacked=True,color=['cyan','hotpink'],edgecolor='black')


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')

plt.legend(['dataset1','dataset2'])
plt.show()
```

```python
sns.histplot([data1,data2],bins=30,kde=True,color=['cyan','hotpink'])


plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Stacked Histogram')

plt.legend(['dataset1','dataset2'])
plt.grid(":")
plt.show()
```

# pie Charts Plotting

```python
# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
        'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]

plt.pie(data,labels=cars)
plt.show()
```

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt


# Creating dataset
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = [23, 17, 35, 29, 12, 41]


# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3, 0.0, 0.0)

# Creating color parameters
colors = ("orange", "cyan", "brown",
		"grey", "indigo", "beige")

# Wedge properties
wp = {'linewidth': 1, 'edgecolor': "green"}

# Creating autocpt arguments


def func(pct, allvalues):
	absolute = int(pct / 100.*np.sum(allvalues))
	return "{:.1f}%\n({:d} g)".format(pct, absolute)


# Creating plot
fig, ax = plt.subplots(figsize=(9, 5))
wedges, texts, autotexts = ax.pie(data,
								autopct=lambda pct: func(pct, data),
								explode=explode,
								labels=cars,
								shadow=True,
								colors=colors,
								startangle=90,
								wedgeprops=wp,
								textprops=dict(color="magenta"))

# Adding legend
ax.legend(wedges, cars,
		title="Cars",
		loc="center left",
		bbox_to_anchor=(1, 0, 0.5, 1))

plt.setp(autotexts, size=8, weight="bold")
ax.set_title("Customizing pie chart")

# show plot
plt.show()
```

> **Nested Pie Chart**

```python
# Import libraries
from matplotlib import pyplot as plt
import numpy as np


# Creating dataset
size = 6
cars = ['AUDI', 'BMW', 'FORD',
		'TESLA', 'JAGUAR', 'MERCEDES']

data = np.array([[23, 16], [17, 23],
				[35, 11], [29, 33],
				[12, 27], [41, 42]])

# normalizing data to 2 pi
norm = data / np.sum(data)*2 * np.pi

# obtaining ordinates of bar edges
left = np.cumsum(np.append(0,
						norm.flatten()[:-1])).reshape(data.shape)

# Creating color scale
cmap = plt.get_cmap("tab20c")
outer_colors = cmap(np.arange(6)*4)
inner_colors = cmap(np.array([1, 2, 5, 6, 9,
							10, 12, 13, 15,
							17, 18, 20]))

# Creating plot
fig, ax = plt.subplots(figsize=(10, 7),
					subplot_kw=dict(polar=True))

ax.bar(x=left[:, 0],
	width=norm.sum(axis=1),
	bottom=1-size,
	height=size,
	color=outer_colors,
	edgecolor='w',
	linewidth=1,
	align="edge")

ax.bar(x=left.flatten(),
	width=norm.flatten(),
	bottom=1-2 * size,
	height=size,
	color=inner_colors,
	edgecolor='w',
	linewidth=1,
	align="edge")

ax.set(title="Nested pie chart")
ax.set_axis_off()

# show plot
plt.show()
```

```python

# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d

# defining all 3 axis
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

fig= plt.figure()
#3d projection
ax=plt.axes(projection='3d')

ax.plot3D(x,y,z,'green')
ax.set_title('3D- Line Plot')
plt.show()
```
