> **3-Dimensional Line Graph**

```python

# importing mplot3d toolkits, numpy and matplotlib
from mpl_toolkits import mplot3d

# defining all 3 axis
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)

fig= plt.figure()
#3d projection
ax=plt.axes(projection='3d')

ax.plot3D(x,y,z,'green')
ax.set_title('3D- Line Plot')
plt.show()
```

> **3-Dimensional Scattered Graph**

```python
# defining axes
z = np.linspace(0, 1, 100)
x = z * np.sin(25 * z)
y = z * np.cos(25 * z)
c = x +y

fig=plt.figure()
ax = plt.axes(projection='3d')

ax.scatter3D(x,y,z,c=c)
ax.set_title('3D-Scatter plot')
plt.show()
```

> **Surface Graphs**

```python
# defining surface and axes
x = np.outer(np.linspace(-2, 2, 10), np.ones(10))
y = x.copy().T
z = np.cos(x ** 2 + y ** 3)

fig=plt.figure()
ax = plt.axes(projection ='3d')

ax.plot_surface(x,y,z,cmap='viridis',edgecolor='cyan')
ax.set_title('3D=surface plot')
plt.show()
```

> **Contour Graphs**

```python
def function(x, y):
	return np.sin(np.sqrt(x ** 2 + y ** 2))


x = np.linspace(-10, 10, 40)
y = np.linspace(-10, 10, 40)

X, Y = np.meshgrid(x, y)
Z = function(X, Y)

fig = plt.figure(figsize=(10, 8))
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, cmap='cool', alpha=0.8)

ax.set_title('3D Contour Plot of function(x, y) = sin(sqrt(x^2 + y^2))', fontsize=14)
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)

plt.show()
```

> **Plotting Mbius strip**

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parameters of the Mbius strip
R = 2

# Define the Mbius strip surface
u = np.linspace(0, 2*np.pi, 100)
v = np.linspace(-1, 1, 100)
u, v = np.meshgrid(u, v)
x = (R + v*np.cos(u/2)) * np.cos(u)
y = (R + v*np.cos(u/2)) * np.sin(u)
z = v * np.sin(u/2)

# Create the plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the Mbius strip surface
ax.plot_surface(x, y, z, alpha=0.5)

# Set plot properties
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Mbius Strip')

# Set the limits of the plot
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([-3, 3])

# Show the plot
plt.show()
```
