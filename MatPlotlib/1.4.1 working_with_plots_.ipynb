# Line Chart

```python
import numpy as np
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
x = np.array([1,2,3,4])
y = x*2

plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel("X_axis")
plt.ylabel("Y_label")
plt.title('Stright line',fontsize=20,color='g')
plt.show()
```

> **line Chart with Annotations**

```python
x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.figure(figsize=(8,6))
plt.plot(x,y,marker='o',c = 'r',ls='--')

for i, (xi, yi) in enumerate(zip(x, y)):
    plt.annotate(f'({xi}, {yi})', (xi, yi), textcoords="offset points", xytext=(0, 10), ha='center')

plt.xlabel("x_axis",fontsize=15)
plt.ylabel('y_axis',fontsize=15)
plt.title("line with markers" ,color='cyan',fontsize=20)

plt.grid()
plt.show()
```

> **Multiple Plots on the Same Axis**

```python
fig, ax = plt.subplots()

x = np.array([1,2,3,4])
y1= x*2

y2 = np.array([4,7,8,9])

plt.plot(x,y1,"--",marker='D')
plt.plot(x,y2,marker='o')

ax.set_xlabel("x_axis")
ax.set_ylabel("y_axis")
ax.set_title("Line Chart")

plt.show()
```

> **Fill the Area Between Two Lines**

```python
plt.plot(x,y1,"--",marker='D')
plt.plot(x,y2,marker='o')

ax.set_xlabel("x_axis")
ax.set_ylabel("y_axis")
ax.set_title("Line Chart")

plt.fill_between(x,y1,y2,color='DarkSalmon',alpha=0.5)

plt.show()
```

> **Fill Between Multiple Lines**

```python
x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2 * np.pi * x)
y2 = 0.8 * np.sin(4 * np.pi * x)

fig, (ax1, ax2, ax3) = plt.subplots(
  3, 1, sharex=True, figsize=(6, 6))

ax1.fill_between(x, y1)
ax1.set_title('fill between y1 and 0')

ax2.fill_between(x, y1, 1)# using fill between
ax2.set_title('fill between y1 and 1')

ax3.fill_between(x, y1, y2)
ax3.set_title('fill between y1 and y2')
ax3.set_xlabel('x')
fig.tight_layout()
```

```python
X = np.linspace(0, 3, 200)
Y1 = X**2 + 3
Y2 = np.sin(X)
Y3 = np.cos(X)

plt.plot(X, Y1, lw=4)
plt.plot(X, Y2, lw=4)
plt.plot(X, Y3, lw=4)

plt.fill_between(X, Y1, Y2, color='k', alpha=.5)
plt.fill_between(X, Y1, Y3, color='y', alpha=.5)

plt.show()
```

# Barplot

> **Stacked bar plot**

```python
N = 5

boys = (20, 35, 30, 35, 27)
girls = (25, 32, 34, 20, 25)
boyStd = (2, 3, 4, 1, 2)
girlStd = (3, 5, 2, 3, 3)
ind = np.arange(N)
width = 0.35

fig = plt.subplots(figsize=(10,7))

p1 = plt.bar(ind, boys,width,yerr=boyStd)
p2 = plt.bar(ind, girls,width, bottom = boys,yerr=girlStd)

plt.xlabel("contribution")
plt.ylabel("contribution by the team")
plt.xticks(ind,('T1','T2','T3','T4','T5'))
plt.yticks(np.arange(0,80,10))
plt.legend(('boys','girls'))
plt.show()
```

> **Create a stacked bar plot**

```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]

plt.bar(x,y1,color='r')
plt.bar(x,y2,color = 'b',bottom=y1)
plt.show()
```

```python
# create data
x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])

plt.bar(x,y1,color='r',label='round1')
plt.bar(x,y2,bottom=y1,color='b',label='round2')
plt.bar(x,y3,bottom=y1+y2,color='g',label='round3')
plt.bar(x,y4,bottom=y1+y2+y3,color='cyan',label='round4')

plt.legend()
plt.xlabel('Team')
plt.ylabel('Score')
plt.title("Score by teams in 4 rounds")
plt.show()
```

> **Annotate Bars in Barplot**

```python
data = {"Name": ["Alex", "Bob", "Clarein", "Dexter"],
        "Marks": [45, 23, 78, 65]}

df = pd.DataFrame(data,columns=['Name','Marks'])
df
```

```python
import seaborn as sns
plot=plt.bar(df.Name,df.Marks)
plt.xlabel("Student Name")
plt.ylabel("Marks")

plt.title("Marks obtained by each student")
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(5,5
                    ))

plots= sns.barplot(x=df.Name,y=df.Marks)

# for annotate we use seaborn
# Iterating over the bars one-by-one
for bar in plots.patches :
   # Using Matplotlib's annotate function and
  # passing the coordinates where the annotation shall be done
  # x-coordinate: bar.get_x() + bar.get_width() / 2
  # y-coordinate: bar.get_height()
  # free space to be left to make graph pleasing: (0, 8)
  # ha and va stand for the horizontal and vertical alignment

  plots.annotate(format(bar.get_height(),'.2f'),
                 (bar.get_x() + bar.get_width()/2,bar.get_height()),
                 ha='center',va='center',size=15,xytext=(0,8),
                 textcoords='offset points')



plt.xlabel("Students", size=14)
plt.ylabel("Marks Secured", size=14)
plt.title("This is an annotated barplot")
plt.show()

```

> **Annotate Bars in Grouped Barplot**

```python
df = sns.load_dataset('titanic')
df
```

```python

# transforming the dataset for barplot
data_df = df.groupby(['sex', 'class']).agg(
    avg_age=('age', 'mean'), count=('sex', 'count'))

data_df = data_df.reset_index()
print(data_df.head())
```

```python
plt.figure(figsize=(5,5))

sns.barplot(x='class',y='avg_age',hue='sex',data=data_df, palette='Greens')

plt.ylabel("Average Age", size=14)
plt.xlabel("Ticket Class", size=14)
plt.title("Simple Grouped Barplot", size=18)
plt.show()
```

> **Annotating the bars**

```python
plt.figure(figsize=(5,5))

plot = sns.barplot(x='class',y='avg_age',hue='sex',data=data_df, palette='Greens')

for bar in plot.patches:
  plot.annotate(format(bar.get_height()),
                (bar.get_x()+bar.get_width()/2.,bar.get_height()),
                ha="center",va='center',xytext=(0,9),textcoords='offset points')




plt.ylabel("Average Age", size=14)
plt.xlabel("Ticket Class", size=14)
plt.title("Simple Grouped Barplot", size=18)
plt.show()
```

**Ceil the values**

```python

plt.figure(figsize=(5,5))

plot = sns.barplot(x='class',y='avg_age',hue='sex',data=data_df, palette='Greens')

for bar in plot.patches:
  plot.annotate(format(round(bar.get_height()),'.0f')+'YEARS',
                (bar.get_x()+bar.get_width()/2.,bar.get_height()),
                ha="center",va='center',xytext=(0,-12),size = 8,textcoords='offset points')




plt.ylabel("Average Age", size=14)
plt.xlabel("Ticket Class", size=14)
plt.title("Simple Grouped Barplot", size=18)
plt.show()
```
