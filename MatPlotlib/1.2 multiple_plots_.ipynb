```python
import numpy as np
import pandas as pd
import matplotlib as pml
import matplotlib.pyplot as plt
```

# Create Multiple Subplots in Matplotlib

```python
# Some data to display
x = [1, 2, 3]
y = [0, 1, 0]
z = [1, 0, 1]

# creating two subplots
fig,ax= plt.subplots(2)

ax[0].plot(x,y)

ax[1].plot(x,z)
plt.show()
```

**Subplots in Matplotlib Using Stacking in Two Directions**

```python
x = np.arange(0.0, 2.0, 0.01)
y = 1 + np.sin(2 * np.pi * x)

# creating subplots
fig,((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(3,2)

ax1.plot(x,y, color="magenta")
ax2.plot(x,y, color="green")
ax3.plot(x,y, color="red")
ax4.plot(x,y, color="orange")
ax5.plot(x,y, color="black")
ax6.plot(x,y, color='blue')

plt.show()
```

> **Multiple Plots Same Figure**

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2)

# Now axs is a 2D array of Axes objects
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 1].scatter([1, 2, 3], [4, 5, 6])
axs[1, 0].bar([1, 2, 3], [4, 5, 6])
axs[1, 1].hist([1, 2, 2, 3, 3, 3, 3, 3, 3, 4, 4, 5])

plt.show()
```

> **Add Title to Subplots in Matplotlib**

```python
# making subplots
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 1].plot(x, x*2)
ax[1, 0].plot(x, x*x)
ax[1, 1].plot(x, x*x*x)

# set titles
ax[0,0].set_title('linear')
ax[0,1].set_title('Double')
ax[1,0].set_title('Square')
ax[1,1].set_title('Cube')

plt.suptitle('main Title',size=20)

plt.show()
```

```python
# another method
x = np.array([1,2,3,4,5])

# making subplots
fig,ax=plt.subplots(2,2)

y = [x,x*2,x*x,x*x*x]
Titles=["Linear", "Double", "Square", "Cube"]

for i in range(4):

  # create subplot
  plt.subplot(2,2,i+1)
  plt.plot(x,y[i])

  plt.gca().set_title(Titles[i])

# set Spacing
fig.tight_layout()
plt.show()
```

**Turn Off the Axes for Subplots**

```python
import matplotlib.tri as mtri

# assign data
x = np.asarray([0, 1, 2, 3, 0.5,
                1.5, 2.5, 1, 2,
                1.5])

y = np.asarray([0, 0, 0, 0, 1.0,
                1.0, 1.0, 2, 2,
                3.0])

triangles = [[0, 1, 4], [1, 5, 4],
            [2, 6, 5], [4, 5, 7],
            [5, 6, 8], [5, 8, 7],
            [7, 8, 9], [1, 2, 5],
            [2, 3, 6]]

# depict illustration
triang = mtri.Triangulation(x, y, triangles)
z = np.cos(1.5 * x) * np.cos(1.5 * y)

fig, axs = plt.subplots()
axs.tricontourf(triang, z)
axs.triplot(triang, 'go-', color ='white')

# turn off the axes
axs.set_axis_off()  # or use ----plt.axis('off')

# assign title
axs.set_title('Triangle illustration')

plt.show()
```

> **Create Different Subplot Sizes**

Create Different Subplot Sizes in Matplotlib using Gridspec

```python
#extra import
from matplotlib import gridspec

fig =plt.figure()

fig.set_figheight(8)
fig.set_figwidth(8)

# create grid for different subplots
spec=gridspec.GridSpec(ncols=2,nrows=2,
                       width_ratios=[2,1],wspace=0.5,
                       height_ratios=[1,2],hspace=0.5)

# initializing x,y axis value
x = np.arange(0, 10, 0.1)
y = np.cos(x)

# adding subplots
ax0= fig.add_subplot(spec[0])
ax0.plot(x,y)

ax1=fig.add_subplot(spec[1])
ax1.plot(x,y)

ax2 = fig.add_subplot(spec[2])
ax2.plot(x,y)

ax3 = fig.add_subplot(spec[3])
ax3.plot(x,y)



plt.show()
```

> **Create Different Subplot Sizes in Matplotlib gridspec_kw**

```python
# here gridspec is not required


# setting different parameters to adjust each grid
fig,ax= plt.subplots(ncols =2, nrows=2, figsize=(7,7),
                     gridspec_kw= {
                         'width_ratios': [3,3],
                         'height_ratios': [3,3],
                         'wspace':0.4,
                         'hspace':0.4}
                     )

# initializing x,y axis value
x = np.arange(0, 10, 0.1)
y = np.tan(x)

# plot the graphs

ax[0][0].plot(x,y)
ax[0][0].set_title("plot 1")

ax[0][1].plot(x,y)
ax[0][1].set_title("plot 2")

ax[1][0].plot(x,y)
ax[1][0].set_title("plot 3")

ax[1][1].plot(x,y)
ax[1][1].set_title("plot 4")

plt.show()
```

> **Create Different Subplot Sizes in Matplotlib subplot2grid**

```python

# creating grid for sub plots
fig = plt.figure()
fig.set_figheight(6)
fig.set_figwidth(6)


ax1 = plt.subplot2grid(shape=(3,3),loc=(0,0),colspan=3)
ax2 = plt.subplot2grid(shape=(3,3),loc=(1,0),colspan=2)# we can also write without names
ax3 = plt.subplot2grid((3,3),(1,2),rowspan=1)
ax4 = plt.subplot2grid((3,3),(2,0))
ax5 = plt.subplot2grid((3,3),(2,1),colspan=1)

#initializing x,y axis value
x = np.arange(0, 10, 0.1)
y = np.cos(x)

# plotting subplots
ax1.plot(x, y)
ax1.set_title('ax1')
ax2.plot(x, y)
ax2.set_title('ax2')
ax3.plot(x, y)
ax3.set_title('ax3')
ax4.plot(x, y)
ax4.set_title('ax4')
ax5.plot(x, y)
ax5.set_title('ax5')

plt.tight_layout()
plt.show()
```

> **set the spacing between subplots**

```python
# create data
x=np.array([1, 2, 3, 4, 5])

# making subplots
fig, ax = plt.subplots(2, 2)

# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 0].set_title("linear")
ax[0, 1].plot(x, x*2)
ax[0, 1].set_title("double")
ax[1, 0].plot(x, x*x)
ax[1, 0].set_title("Square")
ax[1, 1].plot(x, x*x*x)
ax[1, 1].set_title("Cube")
plt.show()
```

**Set the spacing between subplots**

Example 1: Without using pad

```python

fig, ax = plt.subplots(2, 2)
# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 0].set_title("linear")
ax[0, 1].plot(x, x*2)
ax[0, 1].set_title("double")
ax[1, 0].plot(x, x*x)
ax[1, 0].set_title("Square")
ax[1, 1].plot(x, x*x*x)
ax[1, 1].set_title("Cube")

fig.tight_layout()
plt.show()
```

Example 2: Using pad

```python

fig, ax = plt.subplots(2, 2)
# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 0].set_title("linear")
ax[0, 1].plot(x, x*2)
ax[0, 1].set_title("double")
ax[1, 0].plot(x, x*x)
ax[1, 0].set_title("Square")
ax[1, 1].plot(x, x*x*x)
ax[1, 1].set_title("Cube")

fig.tight_layout(pad=5.0)
plt.show()
```

Using subplots_adjust() method

```python
fig, ax = plt.subplots(2, 2)
# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 0].set_title("linear")
ax[0, 1].plot(x, x*2)
ax[0, 1].set_title("double")
ax[1, 0].plot(x, x*x)
ax[1, 0].set_title("Square")
ax[1, 1].plot(x, x*x*x)
ax[1, 1].set_title("Cube")


# set the spacing between subplots
plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)
plt.show()
```

Using subplots_tool() method

```python
fig, ax = plt.subplots(2, 2)
# set data with subplots and plot
ax[0, 0].plot(x, x)
ax[0, 0].set_title("linear")
ax[0, 1].plot(x, x*2)
ax[0, 1].set_title("double")
ax[1, 0].plot(x, x*x)
ax[1, 0].set_title("Square")
ax[1, 1].plot(x, x*x*x)
ax[1, 1].set_title("Cube")

plt.subplot_tool(targetfig= fig)
plt.show()
# it works on jupyter well
```
