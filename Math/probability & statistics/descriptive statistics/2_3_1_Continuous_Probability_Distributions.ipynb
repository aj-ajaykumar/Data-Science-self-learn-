Unlike discrete random variables, which can only take on a finite number of values, continuous random variables can take on an infinite number of values within a given range. If the random variable is continuous in nature, the probability can be calculated with the help of a function called **Probability Density Function (PDF).**

----
<u>**Probability Density Function:**</u> It is a function that describes the probability distribution of a continuous random variable. The PDF is used to specify the probability of the random variable falling within a particular range of values, without specifying the exact value that it will take. The PDF gives the density of the probability rather than the probability mass.

-----
* **Non-decreasing:** The CDF is a non-decreasing function. As the value of x increases, the value of the CDF can only remain constant or increase.
* **Range:** The CDF ranges from 0 to 1, since the probability of the random variable X being less than or equal to its minimum value is 0, and the probability of X being less than or equal to its maximum value is 1.
* **Continuity:** The CDF is a continuous function, meaning that it has no jumps or discontinuities. This is because the probability that a continuous random variable takes on any specific value is zero.
* **Limits:** The CDF approaches 0 as x approaches negative infinity and approaches 1 as x approaches positive infinity.

Now, with the help of CDF the below questions can be answered.

**What is the probability that the height of a person is lesser than or equal to 160cms?**

````python
from scipy.stats import norm
mean = 160  # average height
std_dev = 10 # deviation of the height
probability_Of_Height_Lesser_Than_Or_Equal_To_160=norm.cdf(160, mean, std_dev)
print('probability Of Height Lesser Than Or Equal To 160:',probability_Of_Height_Lesser_Than_Or_Equal_To_160)
````

The probability of a person height being lesser than or equal to 160 is 0.5. That is there is a 50% chance for a person height measured below or equal to 160cms.

**What is the probability that the height of a person is greater than or equal to 170cms?**

````python
mean = 160  # average height
std_dev = 10 # deviation of the height
probability_Of_Height_Lesser_Than_Or_Equal_To_170 = norm.cdf(170, mean, std_dev)
probability_Of_Height_Equal_To_170 = norm.pdf(170, mean, std_dev)
print(probability_Of_Height_Lesser_Than_Or_Equal_To_170)
print(probability_Of_Height_Equal_To_170)
probability_Of_Height_Greater_Than_Or_Equal_To_170 = 1 - probability_Of_Height_Lesser_Than_Or_Equal_To_170 + probability_Of_Height_Equal_To_170
print('probability Of Height Greater Than Or Equal to 170:',probability_Of_Height_Greater_Than_Or_Equal_To_170)
````

The probability of a person height being greater than or equal to 170 is 0.182 i.e., there is a 18.2% chance for a person height measured above or equal to 170cms.
