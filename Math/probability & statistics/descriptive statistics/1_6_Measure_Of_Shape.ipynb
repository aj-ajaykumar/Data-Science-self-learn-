What if there were a few more households which had much larger incomes? In that case, the mean would shift further away and grow larger than the median. What we observe here is termed `skewness.`

Skewness is the measure of lack of symmetry of the data in comparison with symmetric distribution. In symmetrically distributed data set, the mean and the median coincide.

As observed earlier, when there are unusually large or small values (called outliers), mean is affected by their presence. The shift in the mean away from the median is characterized by the nature of the unusual values.

If there is an unusually large value in the data set, the mean becomes greater than the median, this indicates a right-skewed data. On the other hand, if there are unusually small values, the mean becomes lesser than the median, this indicates left-skewed data.

Skewness of a data set is derived mathematically using the following formula:

Skewness = 3 * (Mean - Median) / (Standard Deviation)

Skewness helps us understand if mean is the right measure of central tendency to use while making a decision about the data. In the case of community 1 and 2, the mean and median are pretty close which indicates that "mean" is an acceptable measure to use here.

The skewness in Python can be computed as follows:

```python
#skewness
from scipy.stats import skew
skew([11730,5461,6655,7484,10242,8547,5521,13207,7598,9421,11563,6995,10531,11525,7770,7094,8339,8375,8808,6802])
```

The nature of data to peak or flatten about the mean can be described using kurtosis.

Kurtosis helps us identify whether the data spread around the mean is high or not.

A higher value of kurtosis indicates that data is largely centred around the meanwhile a lower value of kurtosis indicates that data is not centred around the mean.

## **Summary**

* Measures of shape help us describe the data set better and are based on the distribution of the data.

* Skewness indicates the asymmetry of the data set and is a measure of the presence of extreme values in the data set.

* Skewness helps us identify the right measure of central tendency to use for a given set of data.

* Kurtosis indicates the concentration of data around the mean. A high value of kurtosis indicates that large amount of data is centered around the mean.
