## **Range**



he Range is the simplest measure of dispersion that aims at providing an idea of how much the data varies in the data set.

The Range is expressed as

range = maximum value in a data set - minimum value in a data set

In the example cited earlier(shown below for reference), the range of a number of patients in community 1 is (18-2),  whereas that of community 2 is (12-8).

The range is used to indicating the spread of the data, in our case, this indicates that the data of community 1 is more dispersed than that of community 2.

We can find out the range using the following Python code :

```python
def getrange(numbers):
    print(max(numbers))
    print(min(numbers))
    return max(numbers) - min(numbers)
getrange([5,2,17,8,18])
```

Chebyshev's inequality relates the mean and the standard deviation. It is bound on the probability of finding a value.

By this we mean to say that the probability of finding a value which is closer to the mean is high and the further the value disperses from the mean, the lower the probability of finding it in the data set. Simply put, values closer to the mean have high probability and those which are far, have lesser probability of getting found.

This inequality is expressed as `P(|x - µ| ≥ kσ) ≤  1/k2`

## **Z-Score**

Doctor 1 graduated from University A in the year 2015 with an aggregate of 78% whereas Doctor 2 graduated from University B in the year 2015 with an aggregate of 76%. Both these doctors apply for a post at the government hospital in your city. The recruiter now needs to compare their marks to decide on the candidate to select.

We have the following information with us about the marks of the universities:

* For marks scored in university A, the mean is 72 and the standard deviation is 4.

* For marks scored in university B, the mean is 68 and the standard deviation is 3.

How can we compare data from these different data sets?

Comparison of data from different data sets is done by observing how the data compares against the mean in the data set that it originates from.

This comparison can be performed using z-score, which indicates the distance of the data from the mean in terms of standard deviations.

Example: A z-score of 2 indicates that the data is greater than the mean by 2 standard deviations, while a score of -1.3 indicates that the data is lesser than the mean by 1.3 standard deviations

z-score of a specific value(x) is computed using the formula:

`z = (x - mean) / (standard deviation) =  (x - µ) / σ`

z-score of Doctor 1 is (78-72) / 4 = 1.5, and

z-score of Doctor 2 is (76-68) / 3 = 2.33.

This indicates that Doctor 2 is a more suitable candidate in this scenario.

To calculate the percentage of doctors, we can use percentiles

Percentile represents the value below which a given percentage of observations fall.

In the example, we need to calculate the 80th percentile of the sample i.e. the value below which 80% of the data lies (or the value above which 20% of the data lies).

The nth percentile is identified using the following technique:

1. Order all the data in ascending order

2. Multiply the size of the data set by n (the value obtained is called index)

3. Follow the below steps based on the index obtained:

  1. If the index is not a whole number, round it off to the nearest whole number and identify the value in the data set at the obtained index – this is the percentile value

  2. If the index is a whole number, the percentile value is the average of values at the index and value at index+1

Following this procedure, we find that the 80th percentile value of the given data is 11.5

We can find out the nth percentile by the using the following commands:

```python
import numpy as np
x = [8,12,4,11,15,10,8,7,4,11,6,6,8,7,4,9,16,6,14,6]
np.percentile(x,90)

```

```python
# To find out the quantiles we can use the following commands:


#https://en.wikipedia.org/wiki/Quantile
# Can compute empirical quantiles with mqunatiles
from scipy.stats.mstats import mquantiles
x = [8,12,4,11,15,10,8,7,4,11,6,6,8,7,4,9,16,6,14,6]
mquantiles(x,prob=[0.25, 0.5, 0.75])
```

 ## **Quartiles - Part 2**

Quartiles are 3 points in the data set that divide the data set into 4 equal groups.

The first quartile (Q1) is the 25th percentile of the data set i.e. the value below which 25% of the data lies.

The second quartile (Q2) is the 50th percentile of the data set i.e. the value below which 50% of the data lies. This is the median (m) of the data set

The third quartile (Q3) is the 75th percentile of the data set i.e. the value below which 75% of the data lies.

## **Interquartile range**

The interquartile range is the difference between the third quartile(Q3) and the first quartile(Q1). This range is used to indicate the spread of the data and is a better measure than range because min and max (used in range) are highly volatile in the presence of unusual values.

The interquartile range of the example cited earlier is

`IQR = (Q3-Q1) = 11 – 6 = 5`

To calculate the interquartile range in Python we can use the following commands:

```python
#interquartile range
from scipy.stats import iqr
x = [8,12,4,11,15,10,8,7,4,11,6,6,8,7,4,9,16,6,14,6]
iqr(x)
```

## **Summary**

* Measures of dispersion are used to describe the variability of the data.

* Measures of dispersion include range, variance, standard deviation and interquartile range.

* The range is a simple measure used to understand the spread of the data.

* Variance and standard deviation describe the variability of the data with respect to the mean.

* The interquartile range is a more robust measure used to describe the range of the data and is used because range is susceptible to extreme values in the data set.
