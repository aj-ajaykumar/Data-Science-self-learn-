Suppose the permissible error margin in the weight of a single scoop of ice-cream that is being served is 10%, then calculate the probability of weight falling between 101.7 and 124.3 grams for a single scoop of ice-cream.

Using Continuous Uniform Distribution, the probability of the same can be determined.

<u>**The Experiment:**</u>

The experiment is determining the likelihood of the weight of a single scoop of ice-cream assuming any value in the interval [101.7 grams,124.3 grams].

<u>**The Random Variable (X):**</u>

Let, X (random variable)= Weight of a single scoop of ice-cream

X can be any value between 101.7g and 124.3g.

<u>**PMF and CDF:**</u>

The single ice-cream scoop's weight, which is being measured is equally likely to take any values between 101.7g and 124.3g.

Therefore, we can calculate X's PMF as:  

```
PMF, f(x) = 1/(b-a) = 1/(124.3 - 101.7)

                                 = 1/ 22.6

                                 = 0.0442477876106195 ≈ 0.04423

```

Thus, X is a continuous uniform distribution with probability 0.04423 or 4.423%, for all values between 101.7g and 124.3g.

For x = 103,

```
CDF, F(x) = (x-a)/(b-a) = (103-101.7)/(124.3-101.7)

                                      = 1.3/22.6

                                      = 0.0575221238938053 ≈ 0.0575

```


Thus, the probability of weight being 105g for the ice-cream scoop is 0.146 or 14.6%.

The same is computed using python programming as manual computation is not always possible.

The different Python functions present in scipy.stats module associated with Uniform Distribution are as follows:

PDF: uniform.pdf(x, loc, scale)

CDF: uniform.cdf(x, loc, scale)

where,

x - value between the given range

loc - the lower limit of distribution

scale - difference between upper and lower limit

````python
#Continuous Uniform Distribution: PDF
from scipy.stats import uniform
x= 105.0 #change the values to any value in the interval [101.7, 124.3] and observe the output
lower= 101.7 #lower limit of distribution
upper=124.3
diff= upper-lower
#The value of x can be any value in the interval [101.7, 124.3].
#As an example x is taken as 105 in the code shown below.
dunif = uniform.pdf(x,loc = lower,scale = diff)
dunif
````

The below code demonstrates the computation and graphical representation of PDF and CDF of Continuous Uniform Distribution using python.

````python
from matplotlib import pyplot as plt
import numpy as np
from scipy.stats import uniform
lower_limit = 374.56
upper_limit = 424.32
diff=upper_limit-lower_limit
randVar = np.linspace(373,425,200)  # value to calculate PDF for
# calculate PDF using the uniform.pdf() function
pdf = uniform.pdf(randVar, loc=lower_limit, scale=diff)
plt.rcParams['figure.facecolor'] = 'lightblue'
plt.plot(randVar,pdf, color='red')
plt.fill_between(randVar, pdf, color='lightpink')
plt.title('Continuous Uniform - Stock Price scenario PDF',fontsize=15)
plt.xlabel('Random Variable X ',fontsize=13)
plt.ylabel('Probability',fontsize=13)
````

````python
from scipy.stats import uniform
lower_limit = 374.56
upper_limit = 424.32
diff=upper_limit-lower_limit
randVar = np.linspace(373,425,200)  # value to calculate PDF for
# calculate CDF using the uniform.cdf() function
cdf = uniform.cdf(randVar, loc=lower_limit, scale=diff)
plt.rcParams['figure.facecolor'] = 'lightpink'
plt.plot(randVar,cdf, color='blue')
plt.fill_between(randVar, cdf, color='lightblue')
plt.title('Continuous Uniform - Stock Price scenario CDF',fontsize=15)
plt.xlabel('Random Variable X ',fontsize=13)
plt.ylabel('Probability',fontsize=13)
````

The below code demonstrates the computation and graphical representation of PDF and CDF of Normal Distribution using python.

````python
from scipy.stats import norm
mean = 130.5  # mean of the normal distribution
std_dev = 7.9  # standard deviation of the normal distribution
x = np.linspace(100,160,200)  # value to calculate PDF for
# calculate PDF using the norm.pdf() function
pdf = norm.pdf(x,mean, std_dev)
plt.rcParams['figure.facecolor'] = 'lightblue'
plt.plot(x,pdf, color='red',label="mean=130.5, sd=7.9")
plt.fill_between(x, pdf, color='lightpink')
plt.title('Normal Distribution - Graphical representation of PDF',fontsize=15)
plt.xlabel('Random Variable X ',fontsize=13)
plt.ylabel('Probability',fontsize=13)
#plt.xticks(np.arange(100,160))
plt.legend()
````

<u>**Degrees of Freedom:**</u> The shape of the t-distribution depends on the degrees of freedom (df) parameter. The degrees of freedom represent the sample size minus one, and as the sample size increases, the t-distribution approaches the normal distribution.

<u>**Standardization:**</u> The t-distribution is often used in hypothesis testing and confidence intervals, where it is standardized by subtracting the sample mean from the population mean and dividing by the standard error. The resulting standardized value follows a t-distribution with df degrees of freedom.

<u>**PDF and CDF:**</u> The probability density function (PDF) and cumulative distribution function (CDF) of the t-distribution are given by mathematical formulas that involve the gamma function and the incomplete beta function. These formulas can be implemented in software such as Python or R to compute the PDF and CDF for a given value of x and df.

<u>**Confidence Intervals:**</u> The t-distribution is used to compute confidence intervals for the mean of a population when the population standard deviation is unknown. The confidence interval is based on the t-distribution with n-1 degrees of freedom, where n is the sample size.

<u>**Hypothesis Testing:**</u> The t-distribution is used in hypothesis testing to determine whether a sample mean is significantly different from a hypothesized population mean. The test statistic is computed by standardizing the sample mean using the t-distribution, and the p-value is calculated from the t-distribution with n-1 degrees of freedom.

<u>**Robustness:**</u> The t-distribution is considered robust to outliers and non-normality, which means that it can still provide accurate results even when the data are not normally distributed or contain outliers.

<u>**Applications:**</u> The t-distribution is widely used in many fields, including biology, engineering, finance, and social sciences. It is used to test hypotheses about population means, estimate confidence intervals for population means, and perform statistical inference on small samples.
