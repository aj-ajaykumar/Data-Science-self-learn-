## **Law of Large Numbers**

Have you ever thought how an insurance company makes profit by insuring the people? Insurance companies, provides insurance to large number of people who can pay monthly premium, but it ensures that, just a small number of people claims the insurance payment for large, unexpected expenses. For example, 5000 people are insured by paying some premium resulting in large amount of profit to the insurance company but only 50 people may need company to cover the unexpected expenses to some accidents which results in some loss to company. The amount of profit to loss is small because, company ensures that large number of people are paying the premium.
This business model is perfect instance of Law of Large Numbers -LLN.

The Law of Large Numbers states that, as the sample size of random variable increases, their sample mean approaches their theoretical mean (i.e., population mean).

The Law of Large Numbers (LLN) is a theorem that describes the result of performing the same experiment a large number of times. According to this law, the average result obtained from a large number of trials should be close to the expected value and will tend to be closer with more trials.

Suppose X1, X2, ..., Xn are independent random variables with the same underlying distribution. In this case, we say that the random variables Xi (where i can take 1, 2â€¦n) are independent and identically distributed (i.i.d.). In particular, Xi will have the same mean Î¼ and standard deviation Ïƒ.

To explain the Law of Large Numbers in a simpler way, let us generate random numbers using a continuous distribution (i.e., Normal Distribution).

We are generating random numbers using the norm.rvs() function, mean 50 and standard deviation 5 (follow step 1-3 as shown below).

Random numbers mean is calculated using np.mean() function. With the increasing sample size, the mean is approximating to population mean (i.e., 50).

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import norm
# Generating 5 random number
norm.rvs(loc = 50, scale = 5, size = 5)
````

````python
#Law of Large Numbers
print("Mean when sample size is 5: ",np.mean(norm.rvs(loc = 50, scale = 5, size = 5)))
print("Mean when sample size is 10: ",np.mean(norm.rvs(loc = 50, scale = 5, size = 10)))
print("Mean when sample size is 100: ",np.mean(norm.rvs(loc = 50, scale = 5, size = 100)))
print("Mean when sample size is 1000: ",np.mean(norm.rvs(loc = 50, scale = 5, size = 1000)))
print("Mean when sample size is 10000: ",np.mean(norm.rvs(loc = 50, scale = 5, size = 10000)))
print("Mean when sample size is 100000: ",np.mean(norm.rvs(loc = 50, scale = 5, size = 100000)))
````

## **Central Limit theorem**

Accordingly, to LLN, the sample mean will approximate to the population mean with increasing sample size.

But what about the distribution of the sample means?

Is normal distribution followed by the sample means from a population?

Let us try to learn this with an example.

To demonstrate CLT, let us generate random variables from any discrete or continuous distribution, with different sample sizes and plot the histogram of the arithmetic mean and see if it converges to the normal distribution when the sample size increases.

We will use binomial distribution (i.e. np.random.binomial()) in example.

> **Example - Central Limit Theorem(CLT)**

Let with 0.6 probability, the sample size be 10.

We can say that the histogram of random variable means from the binomial distribution (np.random.binomial()) is not a good approximation of normal distribution.



````python
#Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
#With Sample Size 10
xbar = [None]*1000
for i in range(0,1000):
    x = np.random.binomial(10, 0.6, size = 10)
    xbar[i] = x.mean()
print("xbar_10:",np.mean(xbar))
````

**But, as the sample size increases to 100, (i.e., np.random.binomial()), we can observe, the graph started converging to the normal distribution.**

````python
#With Sample Size 100
xbar = [None]*1000
for i in range(0,1000):
    x = np.random.binomial(10, 0.6, size = 100)
    xbar[i] = x.mean()
print("xbar_100:",np.mean(xbar))
````

````python
plt.hist(xbar,bins = 'auto', rwidth = 0.5, color = 'blue')
plt.xlabel('xbar', fontsize=13)
plt.ylabel('Frequency', fontsize=13)
plt.title('n=100', fontsize=15)
plt.show()
````

## **Normality Tests**

Normality tests are important in data analysis because many statistical tests, such as t-tests, ANOVA, and regression analysis, assume that the data being analyzed are normally distributed. If the data is not normally distributed, these tests may give inaccurate or misleading results. Normality tests help ensure that the data meet the assumption of normality and can help identify whether the data needs to be transformed. By performing a normality test, researchers/data analysts can have more confidence in the results of statistical analysis and make more accurate conclusions based on the data.

The Shapiro-Wilk test is a widely used normality test that assesses whether a sample comes from a normal distribution. The null hypothesis is that the sample comes from a normal distribution.

To perform the Shapiro-Wilk test in Python, we can use the shapiro function from the SciPy library. Here's an example on how the same can be done using python.

````python
from scipy.stats import shapiro
data = sample1
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05 # Level of Significance
if p > alpha:
   print('Sample looks Gaussian (fail to reject H0)')
else:
   print('Sample does not look Gaussian (reject H0)')
````

In this example, the data is passed to the shapiro function, which returns the test statistic (stat) and probability value (p) of the test. We then compare the p-value to our chosen significance level (alpha) and interpret the result accordingly.

````python
from scipy.stats import shapiro
data = sample2
stat, p = shapiro(data)
print('Statistics=%.3f, p=%.3f' % (stat, p))
alpha = 0.05 # Level of Significance
if p > alpha:
   print('Sample looks Gaussian (fail to reject H0)')
else:
   print('Sample does not look Gaussian (reject H0)')
````

The Anderson-Darling test is another normality test that assesses whether a sample comes from a particular distribution (in this case, the normal distribution). The test returns a critical value at different levels of significance for a given sample size and compares the sample's test statistic to the critical value.

To perform the Anderson-Darling test in Python, the anderson function from the SciPy library can be used. Here's an example on the same.

````python
from scipy.stats import anderson
data = sample1
result = anderson(data, dist='norm')
print('Statistic: %.3f' % result.statistic)
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
       print('%.1f%%: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
    else:
       print('%.1f%%: %.3f, data does not look normal (reject H0)' % (sl, cv))
````

In this example, the data is passed to the anderson function, along with the distribution we want to test against ('norm' for the normal distribution). The function returns a Result object, which contains the test statistic (result.statistic) and the critical values at different significance levels (result.critical_values). We then loop through the critical values, compare the test statistic to each one, and interpret the result accordingly.

````python
data = sample2
result = anderson(data, dist='norm')
print('Statistic: %.3f' % result.statistic)
for i in range(len(result.critical_values)):
    sl, cv = result.significance_level[i], result.critical_values[i]
    if result.statistic < cv:
       print('%.1f%%: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
    else:
       print('%.1f%%: %.3f, data does not look normal (reject H0)' % (sl, cv))
````

The KS test is a non-parametric test that is used to determine whether a sample comes from a specific distribution or not. In the case of testing for normality, the KS test compares the empirical CDF of the data to the theoretical CDF of a normal distribution with the same mean and standard deviation as the data. The test statistic is the maximum absolute difference between the two CDFs, and the p-value is obtained from a table or using software.

To perform the KS test in Python, the kstest function from the scipy.stats module can be used. Here's an example on the same.

````python
from scipy.stats import kstest, norm
import numpy as np
data = sample1
# Perform KS test
stat, pval = kstest(data, 'norm')
# Print results
print('KS Test Statistic:', stat)
print('p-value:', pval)
alpha=0.05
if pval > alpha:
   print('Sample looks Gaussian (fail to reject H0)')
else:
   print('Sample does not look Gaussian (reject H0)')
````

````python
from scipy.stats import kstest, norm
import numpy as np
data = sample2
# Perform KS test
stat, pval = kstest(data, 'norm')
# Print results
print('KS Test Statistic:', stat)
print('p-value:', pval)
alpha=0.05
if pval > alpha:
   print('Sample looks Gaussian (fail to reject H0)')
else:
   print('Sample does not look Gaussian (reject H0)')
````
