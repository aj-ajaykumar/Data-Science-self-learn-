## **Random Variable and Probability**

As mentioned previously, since machine learning tasks deal with uncertainty on the sampled data, this section introduces basics of probability. The whole concept of probability is based on Experiment and few associated terminologies which are defined as below with few examples:

---

> **Experiment:** An experiment is a procedure or an operation whose result is not predetermined.

Experiment’s results are classifiable.
Experiments are repeatable.
Experiment’s results are definite.
Denoted as “E”

---
> **Outcome:** For a given experiment E, each of the possible result is an outcome.

---
> **Sample space:** Set of all possible outcomes of a given experiment E.

---
> **Event:** The result or Outcome or Observation of a random experiment. Event is a subset of sample space

---
> **Trial:** A single event or experiment with a known outcome. Sometime Trial and Experiment are used synonymously.

---
> **Probability:** It is a hypothetical measurement of the possibility of an event occurring.

It is a number between ‘0’ and ‘1’, where ‘0’ means that the event is impossible, and ‘1’ means that the event is certain to occur.
P(E) – denotes the Probability of an experiment- E.
P(E) - ratio of the number of outcomes to the total number of outcomes which can be a fraction, decimal, or percentage.

Consider an experiment of random selection of three semiconductor chips for a quality check. Suppose

Event of interest: E= All the outcomes in which one or more chips are defective
P(E) = 7/8 = 0.87. i.e., 87% probability that 1 or more chips will be defective.
Similarly, the probability may be estimated for any random experimentation. Is there a way to quantify the probability of an event so that it is easy to understand the characteristics of the sample data?

Yes, it is possible to give a numerical description or quantify the outcome of an experiment called as Random Variable.

---
Random Variable: A real-valued function, defined over a sample space.

---

* Only one real value is assigned by function to each individual outcome.
* X or any other uppercase letter denotes a random variable.
* A lowercase letter, x denotes the real value that can be mapped with a random variable
* x is not a variable like in Algebra. e.g., x+2 = 7, where x value is unknown.
* X is a function and can be depicted as follows:

X = {x1, x2, x3, ....} or X = xk, where k = 1,2,3, …

All the possible outcomes are variable values, and each variable value has a probability. The variable values together with their probabilities are a probability distribution.

---
Probability Distribution: A Mathematical function that describes the probabilities of different outcomes or values of a random Variable.


---
* It specifies the possible values that a random variable can take along with their associate probabilities.
* It is represented through either table or graph like histogram where, X-axis takes random variable value and Y-axis indicate its associated probability.
* Sum of all probabilities is 1.
* All outcomes are assigned equal probabilities
