 ## **Introduction to Sampling**

> **Challenge 1: How many students to select for the surprise test?**

Suppose the surveying agency decides to conduct the surprise test for a selected set of students. Now they must decide the number of students who should appear for the surprise test, such that they represent the performance of all the students in the country.

In statistical terms, the surveying agency must determine the Sample Size.

The sample size is determined by the level of accuracy desired. Level of accuracy indicates how close the parameter estimation is required to be with the population parameters.

The level of accuracy is measured in terms of:

* confidence level

* maximum permissible error

Let us see the effect of these parameters on the sample size.

## **Determining Sample Size**

> **Factors Affecting Sample Size: Confidence Level**

The surveying agency estimates the quality of primary school education in the country by analysing the performance of a selected set of students in the surprise test.

If the agency gives a single point estimate of the performance of students i.e. the average performance of primary school students in the country as 72%, it may not be equal to the actual value.

Instead, if the agency proposes an interval i.e. the average performance of primary school students in the country lies between 70% to 75%, there is a greater certainty associated with this estimate. This interval is called Confidence Interval and the level of certainty associated with it is characterized by Confidence Level.

Confidence level  represents the probability that the population parameter lies within the confidence interval. 95% confidence level implies that the probability of the population parameter being within this interval is 0.95.

To associate a greater certainty i.e. higher confidence level with the estimations, you must analyse more and more samples to represent the population.

Thus, if you associate a higher level of confidence to the analysis you may need to increase the sample size.



> **Factors Affecting Sample Size: Maximum Permissible Error**

Estimated parameters are different from the population parameters because only a sample is considered, not the entire data.

Accurate analysis implies that the difference between the population parameters and the estimated ones are small.

This difference is restricted by specifying maximum permissible error, which is the maximum amount of error that can be accommodated in the analysis.

To reduce the maximum permissible error in the analysis, you may have to analyse more samples.

## **Sampling Techniques**

> **Challenge 2: Which students to select for the surprise test?**

To choose 1120 students from various schools across the country, the agency has two options.

<u>**Option 1:**</u> To randomly choose 1120 students. For example: Conduct a lottery and randomly pick the registration numbers of 1120 students out of all the students from the entire country.

This type of sampling technique is called Simple Random Sampling.

<u>**Option 2:**</u> To choose 1120 students from different schools across states considering factors like population of states, no. of schools in each state etc.

This type of sampling is called Stratified Sampling.

Let us learn more about these sampling techniques.

> **Simple Random Sampling**

Assume that from a population of size N, you create all possible samples of size n. If each of them has an equal probability of getting selected, then the process is called random sampling.

From a random sample, you can understand that:

* A random sample of size n is a set of n observations selected from a Finite/Infinite population

 * Example for finite population: Number of students in a country

 * Example for infinite population: Number of times a coin can be tossed

* Assume that you have a finite population and the size of the population is N. Further assume that the probability of selection of each subset of size n is equal

* A random sample of size n (X1, X2, …, Xn), selected from infinite population having PDF f(x) has each Xi as a random variable whose distribution is given by f(x) and the n random variables are independent

> **Simple Random Sampling in Python**

There are various ways to get a simple random sample and the most common method is called lottery method.

Each of the N observations (population) is assigned a unique number (s1, s2, …, s15) and the numbers are placed in a bowl and shuffled for a blind-folded person to select n numbers.

Let us see how to generate a sample in Python.

```python
#Generating 25 states
states = ["s" + str(i) for i in range(0,16)]
print(states)
```

In Python, you can use choice() function of NumPy library to select n objects randomly among N given observations, as shown below:

```python
#Importing the NumPy library
import numpy as np
#Choosing 8 Random states from ‘states’ without repetition
np.random.choice(states, size=8, replace=False)
```

> **Drawbacks of Simple Random Sampling**

In simple random sampling, the selection is completely based on chance. This way you may select a sample that may not represent the behavior of the entire population.

For example, the 1120 students picked by the agency may all be high performers coincidentally. Then, the performance of these students may not represent the actual quality of primary education. It may give a biased result and such a sample is called a Biased Sample.

Bias refers to the tendency of a statistic to systematically overestimate or underestimate a population parameter.

To avoid such problems, the agency decides to use stratified sampling.

> **Stratified Sampling**

In stratified sampling:

* The population is divided into groups based on certain characteristics. These groups are called Strata
* The samples are chosen randomly from each of these groups

To implement stratified sampling, the agency:

* Selects eight states among the 16 states in the country based on their population

Total schools = 7 * 8 = 56 schools

* Selects 20 students from each school including high, average as well as low performers

Total students = 56 * 20 = 1120 students

<u>**Solution to Challenge 2:**</u> The agency uses the stratified sampling technique to identify 1120 students from the country who are going to appear for the surprise test.



> **Stratified Sampling in Python**

Stratified sampling can be performed in Python using the sample() function, available in Python. Let us understand it using an example.

Suppose you have a list of 12 employees along with their department and job level information.


<center>

image.png

</center>

You can sample the data by grouping it based on department and job level. There are two departments (D1 and D2) and two job levels (2 and 3).

Therefore, you can create four strata corresponding to the unique combination of department and job level. The following Python code shows how to sample one element from each stratum

```python
#Importing the required libraries
import numpy as np
import pandas as pd
#Reading the employee data from emp.csv
path = "/content/drive/MyDrive/Inferential stats/emp.csv"
emp = pd.read_csv(path)
#Display the dataframe emp
display(emp)
print("-----------------------------")
#Rearrange the data by grouping it based on department and job level
tab_emp = pd.crosstab(index = emp["Dept"],  columns=emp["Job_Level"], colnames = [''])
#Display the dataframe tab_emp and see unique combinations of department and job level
display(tab_emp)
print("-----------------------------")
#Taking a random sample from the population using the groupby function based on Department and Job Level
emp.groupby(['Dept','Job_Level'], group_keys=False).apply(lambda x: x.sample(1))
```

The output of crosstab() function shows that there are four unique combinations of department and job level and each combination has three observations.

You can observe that output sample has all combinations of 'Dept' and 'Job_Level'.
