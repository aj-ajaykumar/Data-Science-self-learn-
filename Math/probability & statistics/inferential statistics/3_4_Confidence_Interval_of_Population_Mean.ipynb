> **Confidence Interval of Population Mean**

The surveying agency now intends to find the confidence interval of population mean with 95% confidence level.

The 'sampling distribution of means' is normally distributed as the agency is dealing with a much larger sample size.

Let us see how the agency locates two points (p1, p2) on the sampling distribution curve, such that 95% of the total area within the curve is enclosed. The same is shown in the picture below.

You can see that:

* point p1 encloses 2.5% of the total area under the curve to its left.

* point p2 encloses 95+2.5 = 97.5% of the total area under the curve to its left.

As the distribution is normal, Python function norm.ppf() can be used to find how far points p1 and p2 are from the mean of the distribution in terms of standard deviations.

```python
#Importing the norm function from the stats module of scipy library
from scipy.stats import norm
#Finding confidence interval of population mean with 95% confidence level
p1 = norm.ppf(0.025)
p2 = norm.ppf(0.975)
print("P1 lies",p1,"standard deviations away from the mean")
print("P2 lies",p2,"standard deviations away from the mean")
```
