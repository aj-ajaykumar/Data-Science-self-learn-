###  **Step 2.2: Calculating the P Value**

Since the agency is performing a one-sided hypothesis test, it finds the probability of getting a sample mean -1.33 standard deviations away from the mean of the sampling distribution. The same can be found using the Python function norm.cdf().

```python
#Importing the norm function from the scipy.stats module
from scipy.stats import norm
#Finding the probability of getting a value that is -1.33 standard deviation from mean using norm.cdf() function
print("The probability of getting a value that is -1.33 standard deviation from mean :",norm.cdf(-1.33))
```

The output of the norm.cdf() function, also called as P value, is 0.0917. The output of norm.cdf() function gives the area enclosed under the sampling distribution curve to the left of the point 403.

From the output of norm.cdf() function, it is clear that the probability of getting a sample mean as 403 or less is 0.0917 or 9.17%.

### **Steps 4 & 5: Arriving at the Conclusion**

**Step 4: Concluding whether the Null hypothesis can be rejected or not based on the P value**

The probability of finding a sample mean as 403 or lower is found to be 9.17%.

The level of significance is set as 0.05.

The null hypothesis is rejected if the probability that the observed value will occur is less than 5 percent.

The null hypothesis cannot be rejected, as the probability of getting a sample mean of 403 is greater than 5% (9.17 > 5).

**Step 5: Stating the conclusion**

As the P value > significance level, null hypothesis CANNOT be rejected i.e. based on the performance of selected 100 students, it CANNOT be concluded that the surveying agency's estimation is not valid.

### **Hypothesis Test - Example**

Suppose bulb manufacturer xyz claims that the mean lifetime of a light bulb is greater than 9,500 hours. In a random sample of 30 light bulbs, it is observed that the bulbs last only 9300 hours on average. Assume that the population standard deviation is 110 hours. At 0.05 significance level, test the validity of the xyz manufacturer.

Solution:

**Step 1:**

The claim of the manufacturer: Mean Lifetime > 9500 Hours

Observed Data: Mean Lifetime < 9500

Level of significance: 0.05

**Step 2:**

**Null Hypothesis:** Claim of the manufacturer is correct i.e. the mean lifetime of bulbs is greater than 9500 hours.

**Alternate Hypothesis:** Claim of the manufacturer is NOT correct i.e. mean lifetime of bulbs is less than 9500 hours.

As the alternate hypothesis is about finding that mean is less than 9500 secs, you will be conducting a one-sided hypothesis test.

**Step 3:**

Finding the test statistic as follows:

```python
#Importing required libraries
import numpy as np
from scipy.stats import norm
xbar = 9300 #sample mean
mu0 = 9500 # hypothesized value
sigma = 110 #population standard deviation
n = 30 # sample size
z =(xbar - mu0)/(sigma/np.sqrt(n))  # test statistic
print("Test Statistic:",z)
```

**Step 4:**

Finding the P value as follows:

```python
#Importing the norm function from the scipy.stats module
from scipy.stats import norm
#Finding the p-value
p_value = norm.cdf(z)
print("P-Value :",p_value)
```

**Conclusion: The claim made by the manufacturer is wrong i.e. mean life-time of the bulbs is NOT greater than 9500 hours.**

###  **Hypothesis Test with Known Variance - Example**

Consider the average weight of penguins found in Antarctica last year as 16.3kg. In a sample of 33 penguins in the current year in the same colony, the mean penguin weight is 15.7 kg. Assume the population standard deviation is 2.7 kg. At 0.05 significance level, does the current yearâ€™s mean of penguin weight differ from last year?

**Solution:**

The null hypothesis is that Î¼ = 16.3. Let us begin with computing the test statistic.

```python
#Importing required libraries
import numpy as np
from scipy.stats import norm
xbar = 15.7 #sample mean
mu0 = 16.3 # hypothesized value
sigma = 2.7 #population standard deviation
n = 35 # sample size
z =(xbar - mu0)/(sigma/np.sqrt(n)) # test statistic
print("Test Statistic:",z)
```

The Z statistic is calculated as -1.314684.

Let us now compute the critical values at 0.05 significance level.

```python
#Importing the norm function from the scipy.stats module
from scipy.stats import norm
#Compute the critical values at 0.05 significance level
alpha = 0.05
z_half_alpha = norm.ppf(1-alpha/2)
print("Critical values at 0.05 significance level is",-z_half_alpha,"and",z_half_alpha)
```

The test statistic -1.314684 lies between the critical values -1.9600 and 1.9600. At 0.05 significance level, you can say that the mean penguin weight does not differ from last year i.e. the null hypothesis cannot be rejected.

## **Errors in Hypothesis Testing**

While conducting hypothesis tests, two types of errors can be encountered. These errors are termed as:

* Type I
* Type II

<center>

image.png

</center>


Accepting or rejecting hypothesis is probabilistic. Therefore, you may fail to reject a hypothesis when it is false and reject it when it is true. These error conditions have special names.

Type I error is said to occur when a null hypothesis is rejected when it is true.

Type II error is said to occur when you fail to reject a null hypothesis when it is false.
