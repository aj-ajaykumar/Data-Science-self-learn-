> **Parameter Estimation Techniques**

Parameters are descriptive measures of an entire population. Estimating the population parameter and error associated with it is one of the goals.

Parameter estimation includes two major techniques:

<u>**Point Estimation:**</u> This involves finding a single, most likely value of a parameter and the single value  calculated may be the best estimate of the parameter.

<u>**Interval Estimation (Confidence Interval):**</u>  This involves finding an interval in which the parameter may lie. A confidence level is associated with every confidence interval which suggests the degrees of certainty with which the population parameter lies in the confidence interval.

> **Parameters to be estimated by agency**

The surveying agency finds the point estimates as well as the confidence interval for various population parameters:

* **Estimating the population mean:** This includes

 * finding the point estimate for population mean

 * finding the confidence interval for population mean

* **Estimating the population variance/standard deviation:** This includes

 * finding the point estimate for standard deviation and population variance

 * finding the confidence interval for standard deviation and population variance

## **Point Estimator of Population Mean**

> **Challenge 3.1: Estimating Population Mean**

The surveying agency starts with estimating the population mean.

* finding the point estimate of population mean i.e. finding the mean marks scored in the surprise test, considering all the students in the country have appeared for the surprise test

* finding the confidence interval of population mean with 95% confidence level i.e. finding an interval such that the mean marks scored by all the students in the country lie in the interval with 95% probability

Let us understand how the surveying agency finds the point estimate of population mean.

> **Point Estimate of Population Mean**

Point estimate of population mean needs to be calculated to estimate population mean.

This can be done in two steps:

<u>**Step 1:**</u> The agency must identify a sample statistic that closely represents the population mean. The sample statistic is called Estimator of the population parameter and its value is called the Estimate of population mean.

<u>**Step 2:**</u> After finding the estimator, the agency must check the quality of the estimator. One of the commonly used measure of the quality is the unbiasedness of the estimator.

Assume that you calculate the expected value of the estimator. Bias of this estimator is the difference between the calculated value and the actual value of the parameter (being estimated). Therefore, when this difference is zero, estimator is said to be unbiased.

Once the agency finds the estimator and verifies its unbiasedness, it is the unbiased estimator of the population parameter.

While performing Step 1, there are various methods to find a sample statistic that closely represents the population parameter. One of them is method of Maximum Likelihood Estimation (MLE).

Therefore, the surveying agency decides to use the method of MLE.

> **Maximum Likelihood Estimation**

In statistics, MLE is one of the methods of estimating the (population) parameters of a statistical model from given data. It deals with finding the likelihood function depending on the model that the data follows and finding the condition under which the likelihood function attains the maximum.

The surveying agency is analysing the attribute 'Mark' which is normally distributed. Thus, the agency must find the condition in which the likelihood function for normally distributed data attains the maximum.

## **Sampling Distribution of Means**

> **Interval Estimate of Population Mean**

The surveying agency has found that sample mean is the point estimator of population mean. The sample mean was calculated to be 413.2759, which is used to estimate the population mean. However, the sample mean may have a different value if the agency would have chosen a different sample. This suggests the sample mean varies as the sample changes.

Suppose the agency would have chosen multiple samples of 1120 students across the country and for each sample, if they would have calculated the sample mean, then it would have found a range of values for the sample mean.  

Thus, instead of specifying a single value of sample mean as the point estimate of population mean, the agency may choose to specify this range of values of sample mean for estimating population mean.

This kind of estimation technique is called Interval Estimation and the range/interval is called the confidence interval.

> **Confidence Interval**

In statistics, confidence interval is a type of interval estimate of a population parameter. It is an observed interval, that frequently includes the unknown value of the parameter of interest, if the experiment is repeated. The confidence level associated with the confidence interval determines how frequently the observed interval contains the parameter.

Remember that the surveying agency decided to set the confidence level at 95%.

This indicates that it is required to find a range of sample means such that the population mean falls within the range with 95% probability.

> **Confidence Interval of Population Mean**

As discussed, confidence interval of population mean is a range of values of sample mean, within which the population mean is expected to lie. The range is identified by analysing multiple samples.

From a population of size N if all possible samples of size n are selected:

* sample mean is calculated for each sample

* the frequency distribution of the sample mean is plotted

The resulting distribution is called 'sampling distribution of sample means' or 'sampling distribution of means'.

Note that,

* each point on the 'lsampling distribution of means' curve represents a sample mean

* the area under sampling distribution curve represents the probability of finding a particular sample mean

Therefore, finding the confidence interval for population mean with a confidence level of 95% is about finding the interval under which 95% of the total area under the 'sampling distribution of means' curve lies.

Thus, to find the confidence interval of population mean, you must plot the 'sampling distribution of means', which is the frequency distribution of all possible sample means of a specific sample size.  

However, the surveying agency has only one sample and there is only one sample mean. Moreover, it is not feasible for surveying agency to analyse all possible samples because of the huge population size and the complexity involved in conducting the surprise test.

Now the question that arises at this point is, how would the agency plot the 'sampling distribution of means' by analysing only one sample. Therefore, you need to understand some of the properties of 'sampling distribution of means' to find the answer to this question.
