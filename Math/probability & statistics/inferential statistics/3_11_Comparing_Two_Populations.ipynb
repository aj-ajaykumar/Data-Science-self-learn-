> **Independent Sample Vs. Paired Data**

While estimating the difference between two population means from the difference in sample means, it should be known whether the two samples are independent of each other or not.

There are two types of samples:

* **Independent Samples:** If a sample drawn from one population has no relationship with/dependency on the sample drawn from another population, then the two samples drawn from two different populations are said to be independent of each other.

* **Paired Data:** If each measurement in one sample is paired with a measurement in another sample, then the two samples are said to be dependent on each other.
Let us understand the concept of independent samples and paired samples with some examples.

**Example 1:** To compare the performances of two different teachers,

* **Independent Sample Case:** Randomly assign 10 students each to Teacher 1 and Teacher 2. Compare the performance of both the groups.
* **Dependent Sample Case:** First assign 10 students to Teacher 1 and then assign the same students to Teacher 2. Compare the change in performance of the same set of students.

**Example 2:** To collect the feedback on two different dishes,

* **Independent Sample Case:** Serve Dish 1 to five people and serve Dish 2 to five other people.
* **Dependent Sample Case:** Serve Dish 1 to five people and Dish 2 to the same people.

> **Plotting Sampling Distribution of Difference in Two Sample Means**

You have learnt how to estimate the mean and standard error of 'sampling distribution of difference in sample means' when the population mean and standard error are unknown.

Let us see how the surveying agency plots the sampling distribution.

The surveying agency finds the ratio of variance in two samples to decide whether it is of equal variance or not. Following Python code is executed for the same.

````python
#Importing required libraries
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.stats import t
#Calculating the variance of the marks scored by both school and computing their ratio
# ddof : “Delta Degrees of Freedom” is the divisor used in the calculation is N - ddof, where N represents the number of elements. By default ddof is zero.
#For the calculation of  sample variance , we divide by N-1. Hence ddof =1
var_school1 = np.var(school1,ddof = 1)
var_school2 = np.var(school2,ddof = 1)
var_ratio = var_school1/var_school2
print("-----------------------------------------")
print("Ratio of variance :",var_ratio)
print("-----------------------------------------")
n1 = len(school1)
n2 = len(school2)
print("n1:",n1)
print("n2:",n2)
# Calculating pool standard deviation
pooled_sd = np.sqrt((((n1-1)*var_school1)+((n2-1)*var_school2))/(n1+n2-2))
print("Pool standard deviation:",pooled_sd)
print("-----------------------------------------")
standard_error = pooled_sd * np.sqrt((1/n1)+(1/n2))
print("Standard error",standard_error)
print("-----------------------------------------")
#Ploting the sampling distribution
df = 38
mean_sd  = 28.25
x = np.linspace(-50, 50, 100)
plt.plot(x,t.pdf(x, df))
plt.show()
````

The Python code shows that the ratio of variance is 0.96 which is close to 1. Hence, the surveying agency assumes both the samples have equal variance.

While determining the effect of a factor on two different populations by analysing two samples a hypothesis test called two sample t-test is used.

Assumptions:

* The populations from which the samples were obtained must be normally or approximately normally distributed

* The samples must be independent

* The variances of the populations must be equal

* It is desirable to have groups of the same sample size

Let us know more about two sample t test.

> **Two Sample t Test**

The two sample t test is a hypothesis test, conducted to find whether there is a significant difference between two population means by analysing samples drawn from each population.

This test works when:

* small samples drawn from normally distributed populations are involved. However, it can also be used while dealing with larger samples

* the assumption of normality is violated but only if the underlying distribution is not highly skewed
The surveying agency compares the marks scored by students of the school ST1S1 and ST1S3. To find whether the difference in the population means is statistically significant with a significance level of 0.05, the agency conducts two sample two test.

The agency deals with small samples and it has verified the underlying distribution of marks to be normal distribution. Hence, the assumptions for two sample t test are met.

Since, both the samples are drawn from independent populations, the surveying agency conducts two sample t test for independent samples to determine whether the difference in population means is statistically significant.

**Step 4: Find the critical t value and accept/reject the null hypothesis**

Next step is to find the critical t values based on the level of significance.

For significance level of 0.05, the critical t values are the two points that enclose 95% of the total area under the sampling distribution curve. The area that is enclosed by these two points is called the Acceptance Region. Any point that lies in the acceptance region has more than 5% probability of occurrence. On the other hand, the area outside the acceptance region is called the Rejection Region. Any point that lies in the rejection region has less than 5% probability of occurrence.

**Step 5: Compare the test statistic and the critical t values to arrive at conclusion.**

If test statistic > $t_{critical}$ or test statistic < $-t_{critical}$, then the test statistic lies in the rejection region. Thus, this suggests that, if the null hypothesis is TRUE, there is less than 5% probability of getting the observed difference in mean.

Hence, the null hypothesis is rejected with 0.05 level of significance. So, there is statistically significant difference between the two population mean.

If $-t_{critical}$ <= test statistic <= $t_{critical}$, then the test statistic lies in the acceptance region which suggests that there is more than 5% probability of getting the observed difference in the sample means even if the null hypothesis is true.

Thus, the null hypothesis with 0.05 level of significance cannot be rejected. So, the difference between the population means is statistically insignificant.
